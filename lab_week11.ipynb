The aim of ths exercise is to run a docker container with a python script. Essentially I want you to run the example I showed in the lecture. The Dockerfile should be

<pre>
FROM python:3.9
COPY . /usr/local/src/myscripts
WORKDIR /usr/local/src/myscripts
RUN python -m pip install numpy scikit-learn statsmodels

CMD ["python3", "eg.py"]
</pre>

The python script **eg.py** should be

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])


model = LinearRegression().fit(x, y)

#print(model)

print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```


The docker commands are:

<pre>
docker build -t pythonexample .
docker run pythonexample
</pre>

* The build command creates the docker container
* The run command runs the script

You should be able to store the image on your dockerhub
